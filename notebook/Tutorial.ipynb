# SpaceGeo
国際宇宙ステーション(ISS)の位置情報を取得して、その位置の地上画像を取得する。

- ISSの位置情報取得: [@natronics](https://github.com/natronics)さんの[ISS Location Now](http://api.open-notify.org)を使用
- 地名情報の取得: Googleの[Google Maps Geocoding APIのリバース ジオコーディング](https://developers.google.com/maps/documentation/geocoding/intro?hl=ja#ReverseGeocoding)(緯度経度から住所情報取得)を使用
- 画像情報: [Brendan Viscomi](https://github.com/brendanv/nasa-api)さんのNASA APIのPythonラッパーを使用

```python
import sys

sys.path.append('../spacegeo') 
from spacegeo import SpaceGeo
```

## APIキーの設定
- nasa_api_key: NASA APIキー
    - [NASA API PortalのGet Your API Key](https://api.nasa.gov/index.html#apply-for-an-api-key)で発行したAPIキーを使用
- ggl_geo_api_key: GoogleのGeocoding APIキー
    - [GoogleのGeocoding APIページの「キーの取得、認証」](https://developers.google.com/maps/documentation/geocoding/get-api-key?hl=ja)で発行したAPIキーを使用

設定したAPIでSpaceGeoクラスのインスタンス生成

```python
nasa_api_key = '<NASA API KEY>'
ggl_geo_api_key = '<GOOGLE GEOCODING API KEY>'
sg = SpaceGeo(nasa_api_key, ggl_geo_api_key)
```

## 緯度経度の取得

```python
latlon = sg.getIssLatlon()
print('緯度経度:',latlon)
```

## 緯度経度から位置情報の取得
国名と州の名称を取得

```python
geo_info = sg.getGeoInfo(latlon=latlon)
geo_info
```

## 緯度経度の画像イメージを取得

```python
sg.getEarthImage(latlon=latlon)
```

### OpenCV(cv2)を使っているので、画像のリサイズも可能

```python
sg.getEarthImage(latlon=latlon, size=0.5)
```
